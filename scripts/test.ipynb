```python
data_dir = './../data'
clustering_dir = os.path.join(data_dir, 'Clustering_data')
BASE_URL = '/files/Clustering_data'

records = []

for folder in os.listdir(clustering_dir):
    folder_path = os.path.join(clustering_dir, folder)
    if not os.path.isdir(folder_path):
        continue

    files = os.listdir(folder_path)
    if f"{folder}_centroids.pdb" and f"{folder}.json" in files:
        records.append({
            'uniprot_id': folder,
            'centroids': f"{BASE_URL}/{folder}/{folder}_centroids.pdb",
            'clustering': f"{BASE_URL}/{folder}/{folder}.json"
        })
```

```python
import os
import pandas as pd

data_dir = './../data'
pdb_dir = os.path.join(data_dir, 'PLAAC_PDBs')
BASE_URL = '/files/PLAAC_PDBs'

records = []

for folder in os.listdir(pdb_dir):
    folder_path = os.path.join(pdb_dir, folder)

    records.append({
            'uniprot_id': folder.split(".")[0],
            'pdb': f"{BASE_URL}/{folder}",
        })

# Convert to DataFrame
pdb_df = pd.DataFrame(records)
```

```python
import os
import pandas as pd

data_dir = './../data'
a3d_dir = os.path.join(data_dir, 'aggrescan3d')
BASE_URL = '/files/aggrescan3d'

records = []

for folder in os.listdir(a3d_dir):
    folder_path = os.path.join(a3d_dir, folder)
    if not os.path.isdir(folder_path):
        continue

    files = os.listdir(folder_path)
    if 'A.png' in files and 'A3D.csv' in files:
        records.append({
            'uniprot_id': folder,
            'a3d_png': f"{BASE_URL}/{folder}/A.png",
            'a3d_csv': f"{BASE_URL}/{folder}/A3D.csv"
        })

# Convert to DataFrame
a3d_df = pd.DataFrame(records)
```

```python
def extract_uniprot(folder_name):
    parts = folder_name.split("_")
    return parts[1] if len(parts) > 1 else folder_name

a3d_df['uniprot_id'] = a3d_df['uniprot_id'].map(extract_uniprot)
```
